```python
import haversine as hs   
from haversine import Unit
```

```python
commune_list = ["Anniviers", "Chalais", "Chippis", "Crans-Montana", "Grône", "Icogne", "Lens", "Noble-Contrée", "Saint-Léonard", "Sierre"]
pts_list = [(46.183056, 7.598889), (46.265556, 7.507222), (46.281944, 7.544444), (46.311944, 7.482778), (46.253056, 7.458889), (46.292500, 7.440278), (46.278889, 7.445000), (46.312500, 7.548333), (46.250000, 7.416667), (46.293056, 7.533056)]

len(commune_list), len(pts_list)
```

```python
import pandas as pd
df = pd.DataFrame(index=commune_list)

for i in range(len(commune_list)):
    liste = [hs.haversine(pts_list[i],pts_list[j],unit=Unit.KILOMETERS) for j in range(len(commune_list))]
    df[commune_list[i]] = liste

df
```

```python
warehouse_loc = (46.281778, 7.512944)
liste = [hs.haversine(warehouse_loc,pts_list[j],unit=Unit.KILOMETERS) for j in range(len(commune_list))]
df["Warehouse"] = liste

df
```

```python
df.to_excel("distances.xlsx")
```

```python
alts_list = ["848/1691", "522/1330", "526/532", "577/1546", "503/1064", "1034/1160", "758/1637", "600/907", "500/507", "500/559"]

for i in range(len(commune_list)):
    alt = alts_list[i].split("/")
    
    alt = [int(x) for x in alt]
    df.loc[commune_list[i], "Altitude"] = alt[0]
    df.loc[commune_list[i], "Max Altitude"] = alt[1]
```
